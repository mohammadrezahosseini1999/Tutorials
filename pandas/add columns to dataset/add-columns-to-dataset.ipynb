# Method 2

```python
# Reading dataset
df = pd.read_excel('calculated-WS4.xlsx')
df.sample(6)
```

```python
# we can also use .insert() to adding a new column:
# 2: number of index that we want to new column be in there.
# 'N': the name of new column
# df['A'] - df['Z']: value we want to be in new column

df.insert(2, 'N', df['A'] - df['Z'])
df.sample(2)

# as you see in this method we don't need to shifting new column.
```
