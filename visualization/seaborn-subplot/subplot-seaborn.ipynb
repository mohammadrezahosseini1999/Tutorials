```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt


# we use this library to ignore warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r'C:\Users\Darab Store\Desktop\ML-projects\Git-Hub\Tutorials\visualization\seaborn-subplot\train.csv')
df.head()
```

```python
df.columns
```

```python
df[['Gender', 'family_history_with_overweight', 'FAVC', 'CAEC',
    'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad']] = df[['Gender', 'family_history_with_overweight',
                                                           'FAVC', 'CAEC', 'SMOKE', 'SCC', 'CALC', 'MTRANS', 'NObeyesdad']].astype('category')
df.info()
```

# Continues Varaibles

```python
plt.figure(figsize=(5, 2.5))
for col in df.select_dtypes('float64'):
    sns.displot(data=df, x=col, height=3, aspect=2, kind='kde')
```

# Categorical Varaibles

```python
plt.figure(figsize=(2, 1))
for col in df.select_dtypes('category'):
    sns.countplot(data=df, x=col, palette='deep')
    plt.xticks(rotation=90)
    plt.show()
```
